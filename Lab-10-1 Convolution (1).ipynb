# conv2d's parameters

```python
stride filter를 한번에 얼마나 이동 할 것인가
padding : zero - padding 이미지 상하 좌우에 0의 띠
kernel_size : 잘라서 계산하려는 size
convolution : 합성곱
```

```python
import torch
import torch.nn as nn 
```

# What is convolution? 

convolution - 필터 입력이 뒤집어진 상태로 인풋이 들어감
cross-correlation - 필터 입력이 뒤집지않은 상태로 인풋이 들어감

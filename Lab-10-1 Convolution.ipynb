```python
import torch
import torch.nn as nn 
conv = nn.Conv2d(1,1,7,stride=2,padding=0) # 얘재 2
inputs =torch.Tensor(1,1,64,64)
out = conv(inputs)
out.shape 
```

```python
conv = nn.Conv2d(1,1,5,stride=1,padding=2) # 얘재 3
inputs =torch.Tensor(1,1,32,32)
out = conv(inputs)
out.shape
```

```python
conv = nn.Conv2d(1,1,3,stride=1,padding=0) # 얘재 4
inputs =torch.Tensor(1,1,32,64)
out = conv(inputs)
out.shape
```

```python
conv = nn.Conv2d(1,1,3,stride=1,padding=0) # 얘재 5
inputs =torch.Tensor(1,1,32,64)
out = conv(inputs)
out.shape
```

```python
inputs =torch.Tensor(1,1,28,28)
conv1 = nn.Conv2d(1,5,5)
pool = nn.MaxPool2d(2)
out = conv1(inputs)
out2 = pool(out)
```
